```python
ipl=pd.read_csv('ipl-matches (1).csv')
```

```python
#Find which player has won most pom->in finals and qualifiers
best_player=ipl[~ipl['MatchNumber'].str.isdigit()]
best_player['Player_of_Match'].value_counts()
```

```python
#Toss decision plot
ipl['TossDecision'].value_counts().plot(kind='pie')
```

```python
#how many matches each team has played
team1=ipl['Team1'].value_counts()
team2=ipl['Team2'].value_counts()
(team1+team2).sort_values(ascending=False)
```
