# Series is a 1D and Dataframe is 2D Objects

```python
#can we have multiple index?Let's try
index_val=[('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021)]
a=pd.Series([1,2,3,4,5,6,7],index=index_val)
a
```

```python
#the problems?
# a['cse']
```

```python
#the solution ->multiindex series(also known as Hierarhical Indexing)
#multiple index levels within a single index
```

```python
#how to create multiindex object
#1.pd.MultiIndex.from_turples()
index_val=[('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multi_index=pd.MultiIndex.from_tuples(index_val)
print(multi_index.levels[1])
print(multi_index)
#2.pd.MultiIndex.from_produc()
pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
```

```python
#creating series with multindex object
s=pd.Series([1,2,3,4,5,6,7,8],index=multi_index)
s
```

```python
#how to fetch items from such a series
s['cse',2022]
s['cse']
```

```python
#unstack/make it dataframe
temp=s.unstack()
temp
```

```python
#stack
#make it series
temp.stack()
```

# Working with multiindex data frames

```python
#head and tail
branch_df3.head()
#shape
branch_df3.shape
#info
branch_df3.unstack().info()
#duplicated->isnull
branch_df3.duplicated()
branch_df3.isnull()
```

```python
#Extracting rows-> single
branch_df3.loc[('cse',2022)]
```

```python
#multiple
branch_df3.loc[('cse',2019):('ece',2020):2]
```

```python
#using iloc
branch_df3.iloc[0:5:2]
```

```python
#Extracting columns
branch_df3['delhi']['students']
```

```python
branch_df3.iloc[:,1:3]
```

```python
#Extracting both
branch_df3.iloc[[0,4],[1,2]]
```

```python
#sort index
branch_df3.sort_index(ascending=False)
#both->desending ->diff order
branch_df3.sort_index(ascending=[False,True])
#based on one level
branch_df3.sort_index(level=1,ascending=[False])
```

```python
#multiindex dataframe(col) ->transpose
branch_df3.transpose()
```

```python
#swaplevel
branch_df3.swaplevel()
branch_df3.swaplevel(axis=1)
```

# Long Vs Wide Data

```python
#melt-> single example branch
#wide to long
pd.DataFrame({'cse':[120]}).melt()
```

```python
#melt->branch with year
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt(var_name='Branch',value_name='num_Students')
```

```python
pd.DataFrame(
{
    'branch':['cse','ece','mech'],
    '2020':[100,150,60],
    '2021':[120,130,80],
    '2022':[150,140,70]
}
).melt(id_vars=['branch'],var_name='year',value_name='students')
```

```python
#melt-> real world example
death=pd.read_csv('time_series_covid19_deaths_global.csv')
death
confirm=pd.read_csv('time_series_covid19_confirmed_global.csv')
confirm
```

```python
death=death.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='Date',value_name='num_deaths')
confirm=confirm.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='Date',value_name='num_cases')
```

```python
confirm.merge(death,on=['Province/State','Country/Region','Lat','Long','Date'])[['Country/Region','Date','num_deaths','num_cases']]
```
