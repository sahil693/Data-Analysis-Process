```python
import pandas as pd
import numpy as np
```

```python
courses=pd.read_csv('courses.csv')
student=pd.read_csv('students (1).csv')
nov=pd.read_csv('reg-month1.csv')
dec=pd.read_csv('reg-month2.csv')
matches=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries (1).csv')
```

# Merge

```python
#inner join
student.merge(regs,how='inner',on='student_id')
```

```python
#left Join
courses.merge(regs,how='left',on='course_id')
```

```python
#Right Join
student.merge(regs,how='right',on='student_id')
```

```python
#outer join
student.merge(regs,how='outer',on='student_id').tail()
```

```python
#find total revenue generated
regs.merge(courses,how='inner',on='course_id')['price'].sum()
```

```python
#find month by month revenue
temp=pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp.merge(courses,how='inner',on='course_id').groupby('level_0')['price'].sum()
```

```python
#Print the registration table
#cols->name->course->price
regs.merge(student,on='student_id').merge(courses,on='course_id')[['name','course_name','price']]
```

```python
#Plot bar chart revenue/course
regs.merge(courses,on='course_id').groupby('course_name')[['price']].sum().plot(kind='bar')
```

```python
#Find students who enrolled in both the month
common_student_id=np.intersect1d(nov['student_id'],dec['student_id'])
common_student_id
```

```python
student[student['student_id'].isin(common_student_id)]
```

```python
#find course that got no enrollment
course_id_list=np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(course_id_list)]
```

```python
#find students who did not enroll into any courses
student_list=np.setdiff1d(student['student_id'],regs['student_id'])
student[student['student_id'].isin(student_list)]
```

```python
#Print student name ->partner name for all enrolled student
#Using Self_Join
student.merge(student,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

```python
#find top 3 student who did most number enrollments
regs.merge(student,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
#Find top 3 students who spent most amount of money on courses
regs.merge(student,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
#find top 3 stadium with highest sixes / match ratio
```

```python
temp_df=delivery.merge(matches,left_on='match_id',right_on='id')
six_df=temp_df[temp_df['batsman_runs']==6]
```

```python
#stadium ->sixes
num_sixes=six_df.groupby('venue')['venue'].count()
```

```python
num_matches=matches['venue'].value_counts()
```

```python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
```

```python
#find oranage cap holder of all seasons
temp_df=delivery.merge(matches,left_on='match_id',right_on='id')
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first').sort_values('season')
```
